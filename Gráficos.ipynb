# Histograma

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
histograma = np.histogram(base.iloc[:,1], bins = 6)
```

# Dispersão

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

eixo x = circunferencia
eixo y = volume

```python
plt.scatter(x=base.Girth, y=base.Volume, color="red", marker = '*')
plt.title('Árvores')
plt.ylabel('Volume')
plt.xlabel('Circunferência')
plt.show()
```

```python
plt.plot(base.Girth, base.Volume)
plt.title('Árvores')
plt.ylabel('Volume')
plt.xlabel('Circunferência')
plt.show()
```

# Boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
```

notch = true é a curva de confiança dos dados, em relação à mediana e em relação ao grupos. no caso abaixo, são os dados entre 20 e 30.

```python
plt.boxplot(base.Volume, vert=False, showfliers=False, notch=True)
plt.title('Árvores')
plt.ylabel('Volume')
plt.show()
```

# Gráfico 3D

```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
base3 = pd.read_csv('orchard.csv')
base3.shape
```

```python
base3.head()
```

```python
figura= plt.figure()
eixo = figura.add_subplot(1,1,1, projection='3d')
eixo.scatter(base3.decrease, base3.rowpos, base3.colpos)
eixo.set_xlabel('Decrease')
eixo.set_ylabel('Rowpos')
eixo.set_zlabel('Colpos')
```
